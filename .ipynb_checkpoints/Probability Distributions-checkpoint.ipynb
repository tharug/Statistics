## Random Variable
A _random Variable_ is a variable whose specific outcomes are assumed to arise by chance or according to some random or _stochastic_ mechanism. The chance of observing a specific value, or one within a specific interval, is associated with a probability. The function that defines these probabilities is refered to as a _probability distribution_

**Realizations**  
When observations are made of a random variable, these are referred to as _realizations_

* When we roll a die, the outcomes are numbers 1,2,3,4,5,6
* When the temperature is taken of room, technically, it can be within the range $-\infty < t < \infty$

A discrete random variable is one that can only take on finite or countable infinite set of values.  
A continuous random variable can take on all values in some interval.  
Whether a given random variable is discrete or continuous has consequences for the way in which we think about, or utilize, the probabilities associate.

### Discrete Random Variables
Probability distribution tied to a discrete random variable is called a _probability mass function_. Since these define the probabilities of all possible outcomes, the sum of the probabilities in any probability mass function will always be exactly 1.  

**Cumulative Probability Distribution of Discrete Random Variables**  
The cumulative probability for a random variable X is 'the probability of observing less than or equal to x' and written as $P(X\leq{x})$. The graph of a cdf of a discrete random variable is a step function.

Generally, for a probability mass function;
* There are k distinct outcomes $x_1,x_2,....x_k$
* For each $x_i$, where $i={1,...,k}$, $0 \leq P(X=x_1) \leq 1$
* $\sum_{i=1}^{k}P(X=x_i) = 1$

**Mean and Variance of a Discrete Random Variable**  
For a discrete random variable X, the mean $\mu_x$, also referred to as the expectation or the expected value, is the 'average outcome' that is expected over many realizations.  
For a random variable X with k possible outcomes;
$$\mu_x=\sum_{i=1}^{k}x_iP(X=x_i)$$  

The variance of a random variable is defined by;
$$\sigma^2=\sum_{i=1}^{k}(x_i-\mu_x)^2P(X=x_i)$$


### Continuous Random Variables
Becouse there are infinitly many real numbers within a given interval, the probability of any particular value will be essentially zero, so for a continuous random variable there is no such thing as a point probability, instead we talk about a _probability density_. The density for a continuous distribution is a measure of the probability of “getting a value close to x”. The probability distribution tied to a continuous random variable is called a _probability density function_. 

Probabilities are computed as 'areas underneath' that continuous function, and, as with discrete random variables, the “total area” underneath a continuous probability distribution must evaluate to exactly 1.  

**Cumulative Probability Distribution of Continuous Random Variable**  
The cumulative probability distribution for a continuous variable is interpreted in the same way as for a discrete variable. Given a certain value _w_, the cumulative distribution function provides the probability of observing the value _w or less_. More generally, the cumulative probability for a continuous random variable can be found by calculating the area under the density function of interest from $-\infty$ to w.  

**Mean and Variance of a Discrete Random Variable**  
For a continuous variable $w$ with density function $f$, the mean $\mu_w$ is again interpreted as the 'average outcome' that we would expect over many realizations.
$$\mu_w=\int_{-\infty}^{\infty}w.f(w)dw$$

For $w$, the variance $\sigma^2$,  is given by;
$$\sigma^2=\int_{-\infty}^{\infty}(w-\mu_w).f(w)dw$$

### Bernoulli Random Variable
A Bernoulli random variable has exactly two possible outcomes. Typically labled a 'success' and 'failure'.  
If a bernoulli random variable, takes values 1 (success) with probability _p_ and 0 with probability _(1-p)_, then the mean and the standard deviation of the Bernoulli random variable is given by
$$\mu=p$$

$$\sigma=\sqrt{p(1-p)}$$

```python
def bernoulli_trial(p):
    return 1 if np.random.random() <p else 0
```

```python
bernoulli_rv=[]

for x in range(100):
    result=bernoulli_trial(.3)
    bernoulli_rv.append(result)
```

### Binomial Distribution
Binomial random variable counts the number of successes($k$) in $n$ number of bernoulli trials. For a binomial random variable the probability mass function is given by;
$$f(x)=\frac{n!}{x!(n-x)!}.p^x(1-p)^{n-x}$$
Where $p$ is the probability of success in at each bernoulli trial.

**Conditions for Binomial Distribution**  
1. The trials should be independent
1. The number of trials, $n$, is fixed
1. Each of the trials is a bernoulli trial
1. The probability of success, $p$ is the same for each trial


The mean, variance and standard deviation of the number of observed successes are;
$$\mu=np$$

$$\sigma^2=np(1-p)$$

$$\sigma=\sqrt{np(1-p)}$$

```python
def binomial(n,p):
    k=sum([bernoulli_trial(p) for _ in range(n)])
    return k
```

```python
binormial_rv=[]
for x in range(10000):
    result=binomial(20,0.3)
    binormial_rv.append(result)
```

```python
plt.hist(binormial_rv,bins=np.arange(1,21))
plt.xticks(np.arange(0,20))
plt.xlabel('No. of Successes')
plt.ylabel('No of Occurences')
```

```python
# Using scipy.stats module to initiate a random variable
binormial_rv=stats.binom(20,0.3)

# Plot the probability mass function for the random variable
x=np.arange(0,20)
plt.bar(x,binormial_rv.pmf(x))
plt.xticks(np.arange(0,20))
plt.title('Probability Mass Function for a binomial RV (200,0.3)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```

```python
# Plot the cumulative distribution function for the random variable
x=np.arange(0,20)
plt.bar(x,binormial_rv.cdf(x))
plt.xticks(np.arange(0,20))
plt.title('Cumulative Distribution Function for a binomial RV (200,0.3)')
plt.ylabel('P(X<=x)')
plt.xlabel('x');
```

### Geometric Distribution
The geometic distribution counts the number of failures before the first success is recorded.

```python
def geometric(p):
    trial=1
    # perform trails untill success
    while np.random.random() >=p:
        trial+=1
    return trial

# Perform 1000 experiments and record how many trails were perforn in
# each experiment

geo_rv=[]
for x in range(10000):
    geo_rv.append(geometric(.4))
    
##
plt.hist(geo_rv,bins=np.arange(1,21),edgecolor='w')
plt.xticks(np.arange(0,20))
plt.xlabel('No. of Successes')
plt.ylabel('No of Occurences')
```

```python
# Using scipy.stats module to initiate a random variable
geom_rv=stats.geom(0.4)

# Plot the probability mass function for the random variable
x=np.arange(0,20)
plt.bar(x,geom_rv.pmf(x))
plt.xticks(np.arange(0,20))
plt.title('Probability Mass Function for a geometric RV (0.4)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```

```python
# Plot the cumulative distribution function for the random variable
x=np.arange(0,20)
plt.bar(x,geom_rv.cdf(x))
plt.xticks(np.arange(0,20))
plt.title('Cumulative Distribution Function for a geometric distribution (0.4)')
plt.ylabel('P(X<=x)')
plt.xlabel('x');
```

### Negative Binomial Distribution
The negative binomial distribution counts the number of failures observed before $n$ number of successes are recorded.

```python
def neg_bi(k,p):
    failure=0
    success=0
    # perform trails untill k number of successes have been achieved
    while success < k:
        if np.random.random() < p:    # Success
            success+=1
        else:
            failure+=1
    return failure

# Perform 1000 experiments and record how many failures were perforn in
# each experiment

neg_rv=[]
for x in range(10000):
    neg_rv.append(neg_bi(10,.6))
    
##
plt.hist(neg_rv,bins=np.arange(.5,21.5),edgecolor='w')
plt.xticks(np.arange(0,22))
plt.xlabel('No. of Successes')
plt.ylabel('No of Occurences')
```

```python
# Using scipy.stats module to initiate a random variable
neg_binorm_rv = stats.nbinom(10,0.6)

# Plot the probability mass function for the random variable
x = np.arange(0,20)
plt.bar(x, neg_binorm_rv.pmf(x))
plt.xticks(x)
plt.title('Probability Mass Function for a negative binomial RV (k = 10,p = 0.6)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```

```python
# Plot the cumulative distribution function for the random variable
x=np.arange(0,20)
plt.bar(x,neg_binorm_rv.cdf(x))
plt.xticks(np.arange(0,20))
plt.title('Cumulative Distribution Function for a negative binomial RV (k = 10,p = 0.6)')
plt.ylabel('P(X<=x)')
plt.xlabel('x');
```

## Normal Distribution
The normal distribution model describes a symmetric, unimodal, bell-shaped curve. The shape of the normal distribution can be adjusted using two parameters: the mean and the standard deviation.
Changing the mean shifts the bell curve to the left or right, while changing the standard deviation stretches or contrict the curve.  
When the normal distribution centered on $\mu$ = 0 and a standard deviation of $\sigma$ = 1, is known as the standard normal.

```python
# Probability density function
x=np.arange(-5,5,.01)
pdf=stats.norm.pdf(x,scale=1, loc=0)
plt.plot(x,pdf)

pdf=stats.norm.pdf(x,scale=1, loc=1)
plt.plot(x,pdf)

pdf=stats.norm.pdf(x,scale=2, loc=0)
plt.plot(x,pdf)
```

```python
# cumulative distribution function
cdf=stats.norm.cdf(x,scale=1, loc=0)
plt.plot(x,cdf)
```

**Normal approximation of the binomial distribution**<br>
The binomial distribution with probability of success p is nearly normal when the sample size n is sufficiently large that np and n(1 − p) are both at least 10. The approximate normal distribution has parameters corresponding to the mean and standard deviation of the binomial distribution.

```python
## xs=range(min(binormial_rv),max(binormial_rv))

binomial_rv = stats.binom(50,0.4)

# mean and standard deviation of binormial distribution
bi_mean=binomial_rv.mean()
bi_std=binomial_rv.std()

# plotting the binomial distribution
x=np.arange(5,35,0.5)
plt.bar(x, binomial_rv.pmf(x));

# Plotting normal distribution with parameters for mean and standard deviation
normal_rv=stats.norm(scale=bi_std, loc=bi_mean)
plt.plot(x,normal_rv.pdf(x),color='r')
```
